# Задание повышенного уровня.

Задан объект в пространстве 2D. 
Изменить положение объекта (сместить по оси x на 300 пикселей, по оси y на 200 пикселей).

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(
    figsize=(4, 4),
    )
A = [
    [-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]
    ]
A = np.array(A)
x = A[:,0]
y = A[:,1]
plt.plot(x, y, color='b')

B = np.array(A) + [300, 200, 0]

x = B[:,0]
y = B[:,1]
plt.plot(x, y, color='g')

plt.ylim([-150, 350])
plt.xlim([-120, 410])
plt.legend(['исходный', 'новый'])
print(B)
plt.grid()
```

```python
"Уменьшить объект в два раза. Первый вариант"
plt.figure(
    figsize=(4, 4),
    )
B = np.array(B)
x = B[:,0]
y = B[:,1]
plt.plot(x, y, color='b')

C = np.array(B) / 2
x = C[:,0]
y = C[:,1]
plt.plot(x, y, color='g')
plt.legend(['исходный', 'новый'])
print(C)

plt.ylim([-150, 350])
plt.xlim([-120, 410])
plt.grid()
```

```python
'Второй вариант'
TransformArray = np.array([
    [0.5, 0, 0],
    [0, 0.5, 0],
    [0, 0, 1]
])
C = []
for point in B:
  new_point=np.dot(TransformArray,np.transpose(point))
  C.append(new_point)
C = np.array(C).astype(int)
print(C)
```

```python
plt.figure(
    figsize=(4, 4),
    )
x = B[:,0]
y = B[:,1]
plt.plot(x, y)

x = C[:,0]
y = C[:,1]
plt.plot(x, y, color='g')

plt.ylim([-150, 350])
plt.xlim([-120, 410])
plt.legend(['исходный', 'новый'])
plt.grid()
```

Повернуть на 130 градусов.

```python
from math import cos, sin, pi

theta = 130 * 180/pi
TransformArray = np.array([
    [cos(theta), sin(theta), 0],
    [-sin(theta), cos(theta), 0],
    [0, 0, 1]
])
```

```python
D = []
for point in C:
  new_point=np.dot(TransformArray,np.transpose(point))
  D.append(new_point)
D = np.array(D).astype(int)
print(D)
```

```python
plt.figure(
    figsize=(4, 4),
    )
x = C[:,0]
y = C[:,1]
plt.plot(x, y)

x = D[:,0]
y = D[:,1]
plt.plot(x, y, color='r')

plt.ylim([-250, 350])
plt.xlim([-200, 410])
plt.legend(['новый', 'исходный'])
plt.grid()
```

Выполнить смещение объекта и масштабирование (применить 2 матрицы трансформаций). Смещение на (200, 300) и масштабирование на (0.5, 1,2)

```python
TransformArray1 = np.array([
    [1, 0, 200],
    [0, 1, 300],
    [0, 0, 1]
])
TransformArray2 = np.array([
    [0.5, 0, 1],
    [0, 1.2, 1],
    [0, 0, 1]
])
TransformArray = TransformArray1 * TransformArray2
print(TransformArray)
```

```python
F = []
for point in Q:
  new_point=np.dot(TransformArray,np.transpose(point))
  F.append(new_point)
F = np.array(F).astype(int)
print(F)
```

```python
plt.figure(
    figsize=(4, 4),
    )
x = Q[:,0]
y = Q[:,1]
plt.plot(x, y, color='b')

x = F[:,0]
y = F[:,1]
plt.plot(x, y, color='g')
plt.legend(['исходный', 'новый'])

plt.ylim([-250, 250])
plt.xlim([-250, 200])
plt.grid()
```
